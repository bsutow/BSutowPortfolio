```python
#Setup#
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Prints data#
data= pd.read_csv('/Users/Brett/Downloads/dodgers.csv')
data
```

```python
#Shows all unique and checks for errors#
print("Month:", data['month'].unique())
print("Days of Week:", data['day_of_week'].unique())
print("Opponent:", data['opponent'].unique())
print("Sky:", data['skies'].unique())
print("Day or Night:", data['day_night'].unique())
print("Cap Giveaway:", data['cap'].unique())
print("Shirt Giveaway:", data['shirt'].unique())
print("Fireworks:", data['fireworks'].unique())
print("Bobblehead Giveaway:", data['bobblehead'].unique())
```

```python
#Plots a higogram to show the distribution
plt.rcParams['figure.figsize'] = [12, 10]
bins = [10000,15000,20000, 25000, 30000, 35000, 40000, 45000, 50000, 55000, 60000,65000]
sns.distplot(data.attend, kde = False, bins = bins)
plt.xticks(bins)
plt.show()
```

```python
#Looks at average temperature and game attendance#
x = {'temp':'temperature_average', 'attend':'attendance_average'}
tempdf = data.groupby('month').agg({'temp':'mean', 'attend':'mean'}).rename(columns=x).sort_values('temperature_average')
tempdf
```

```python
plt.scatter(x = 'temp', y = 'attend', data = data)
plt.show()
```

```python
#Plots regression plot to look at attendance data further#
#Looks to see if there is a relationship between weather and game attendance#
sns.regplot(x = 'temp', y = 'attend', data = data)
plt.show()
```

```python
#Box Plot showing attendance per month# 
#Shows a better breakdown of the average attendance per month compared to the max and mins#
sns.boxplot(x = 'month', y = 'attend', data = data)
plt.show()
```

```python
#Shows same breakdown as above in easier to read fashion#
plt.scatter(x = 'month', y = 'attend', data = data)
plt.show()
```

```python
#Bar shows attendance per month total#
plt.bar(x=data['month'], height=data['attend'], color='purple')
plt.xlabel('Month')
plt.ylabel('Attendance')
plt.show()
```

```python
#Shows desciption of month data#
data.groupby('month').attend.describe()
```

```python
#Box Plot showing attendance per month# 
#Shows a better breakdown of the average attendance per month compared to the max and mins#
sns.boxplot(x = 'day_of_week', y = 'attend', data = data)
plt.show()
```

```python
#Shows same breakdown as above in easier to read fashion#
plt.scatter(x = 'day_of_week', y = 'attend', data = data, )
plt.show()
```

```python
#Bar shows attendance per day of week total#
plt.bar(x=data['day_of_week'], height=data['attend'], color='purple')
plt.xlabel('Day of Week')
plt.ylabel('Attendance')
plt.show()
```

```python
#Shows desciption of day of the week data#
data.groupby('day_of_week').attend.describe()
```

```python
#Attendance based off opponents#
sns.boxplot(x = 'opponent', y = 'attend', data = data)
plt.show()
```

```python
#Same breakdown but scatter plot format#
plt.scatter(x = 'opponent', y = 'attend', data = data, )
plt.show()
```

```python
#Describes data by opponent# 
data.groupby('opponent').attend.describe()
```

```python
#Looks at giveaways and day of week in month#
data.groupby(['month','day_of_week','cap','shirt','bobblehead','fireworks']).attend.describe()
```

# Results # 
Looking at the data presented above in our charts. We can see the best night to have a giveaway to boost attendance would be on a Monday in October. This is because Monday's on average are the lowest attended games during the season. Ideally, this game would be in October which is the lowest attended month for the Dodgers. However, this is also based off opponents that the Dodgers play as well. The lowest attended opponent is the Atlanta Braves. On average I would recommend doing a giveaway on a Monday in the month of October. Especially if the team plays a low attended opponent like the Atlanta Braves. In these examples I broke down the means of the data and looked further at the standard deviations as well as the mins and maxs. Through doing this I was able to see which opponents or giveaways created higher attedance during a low attendance time period. In my opinion the Dodgers should focuse on their 3 lowest attended months which are October, May, September pairing that with opponent like the Braves, Astros, or Brewers. These months and opponents are low attended times and teams, which in turn makes the decision process easier for giveaways. These giveaways as shown in the data above help improve attendance and allow for more fans to come to games. Pairing giveaways with typical low attendance games allow for an increase in attendance during tiime periods when fans may not come to games. Which is significant because it is creating attendance during low attedance time periods.
